# **Estadísticas Descriptivas**

Vamos a mirar algunos elementos relacionados con las estadísticas descriptivas utilizando R, es un tema introductorio. Se presentan de igual manera algunas gráficas.

```r
# Abrir archivos CSV

bd <- read.csv(file = "buses.csv", header = TRUE, sep = ";")
head(bd)
```

int = números enteros
dbl = números decimales
chr = caracteres o letras o texto

```r
# Descripción de los datos

str(bd)
```

Algunos de los datos tienen valores NA, esto representa un problema potencial para este caso particular. Las estadísticas descriptivas requieren ser numéricas, por lo tanto, en este caso particular es necesario hacer una transformación. No en todos los casos es necesario, no siempre un NA es lo mismo o es igual a 0, eso dependerá de la necesidad de cada caso.

```r
# Transformación de una columna

bd$COOMICRO.LTDA[is.na(bd$COOMICRO.LTDA)] <- 0

# Transformación de todos los valores

bd[is.na(bd)] <- 0

head(bd)
```

```r
# Promedio
mean(bd$COOMICRO.LTDA)

# Mediana
median(bd$COOMICRO.LTDA)

# Desviación típica

sd(bd$COOMICRO.LTDA)

# Varianza

var(bd$COOMICRO.LTDA)

# Percentiles

quantile(bd$COOMICRO.LTDA)

quantile(bd$COOMICRO.LTDA, 0.50)
```

```r
mean(bd$TRASAN.PLUS.S.A.S)
```

```r
median(bd$TRASAN.PLUS.S.A.S)
```

```r
summary(bd)
```

```r
# Suma de todos los buses

bd$total <- bd$COOMICRO.LTDA + bd$COOPTRANSORIENTE + bd$COOTRANSFRONORTE.LTDA + bd$CORTA.DISTANCIA.S.A. + bd$GUASIMALES.S.A + bd$ONTIVEROS.SAS+
  bd$PETROLEA.S.A + bd$RISARALDA.S.A + bd$TONCHALA.S.A + bd$TRANS.ORIENTAL.S.A. + bd$TRASAN.PLUS.S.A.S + bd$TRASAN.S.A.S

sum(bd$total)

head(bd)
```

```r
bd$periodo <- factor(bd$periodo)

table(bd$periodo)
```

Recordemos, tenemos dos tipos de variables en terminos generales, continuas y discretas. En estadística descriptiva, según sea el caso, podemos trabajar con alguna de estas.

```r
# Resumen variable "Total"

summary(bd$total)
```

```r
hist(bd$total, main = "Histograma total buses por año en Cúcuta y A.M.",
    xlab = "", ylab = "Frecuencia")
```

```r
quantile(bd$total)
boxplot(bd$total)
```

```r
plot(bd$periodo, bd$total,
xlab = "(Años)",
ylab = "Total de buses Cúcuta y A.M.")
```
